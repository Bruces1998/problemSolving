```python
'''
Linked List
'''

class Node:
    
    def __init__(self, data,address,next_add):
        self.data = data
        self.address = address
        self.next = next_add
        
        
    def point(self, next_add):
        self.next = next_add
    
    def __str__(self):
        return "data:{}\taddress:{}\tnext:{}".format(self.data, self.address, self.next)
```

```python
class LinkedList:
    def __init__(self):
        self.root = None
        self.list = []
   
        
    def add_element(self,data,address,next_address):
        if self.root != None:
            node = Node(data, address,next_address)
            self.list.append(node)
            
        else:
            self.root = Node(data, address,next_address)
            self.list.append(self.root)
            
            
    def show(self):
        for i,data in enumerate(self.list):
            print("Node{} is {}".format(i,data))
            
        
```

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        mapp = {'I':1,'II':2,'III':3,'IV':4,'V':5,'VI':6,'VII':7,'VIII':8,'IX':9,'X':10,'L':50,'C':100,'D':500,'M':1000}
        summ=0
        toggle = 0
        if s in mapp.keys():
            return mapp[s]
        s = [i for i in s]
        for i in s[:-1:]:
            if toggle==1:
                toggle=0
                continue
            if mapp[i]< mapp[s[s.index(i)+1]]:
                summ+= mapp[s[s.index(i)+1]] -mapp[i]
                toggle=1
                
            else:
                summ+=mapp[i]
                
        if toggle==0:
            summ+=mapp[s[-1]]
            
        return summ
                
            
```

```python
def zero(arr):
    (n, m) = arr.shape
    rows = np.zeros((n,))
    columns = np.zeros((m,))
    for i in range(n):
        for j in range(m):
            if arr[i][j]!=0:
                continue
                
            else:
                if rows[i]==0:
                    rows[i]=1
                    
                if columns[j]==0:
                    columns[j]=1
                    
    for i in range(n):
        if rows[i]:
            arr[i,:] = np.zeros((n,))
            
    for j in range(m):
        if columns[j]:
            arr[:,j] = np.zeros((m,))
            
            
    return arr
```

```python
def oneAway(s1,s2):
    if s1==s2:
        return True
    sum1=0
    sum2=0
    for i in range(len(s1)):
        sum1+=(ord(s1[i])*(i+1))
        
    for i in range(len(s2)):
        sum2+=(ord(s2[i])*(i+1))
        
        
    for i in range(len(s2)):
        tempsum=0
        temp = s2[:i] + s2[i+1:]
        for j in range(len(temp)):
            tempsum +=(ord(temp[j])*(j+1))
            
        if tempsum==sum1:
            return True
        
        
    return False      
```

```python
def oneInsertAway(s1, s2):
    index1=0
    index2=0
    while(index2<len(s2) and index1<len(s1)):
        if s1[index1]!=s2[index2]:
            if index1!=index2:
                return False
            index2+=1
            
        else:
            index1+=1
            index2+=1
            
    return True
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.top = None
        self.end = None
        
    def add(self,data):
        node = Node(data)
        
        if self.top == self.end == None:
            self.top = self.end = node
        else:
            self.end.next = node
            self.end = node
            
            
    def print(self):
        a = self.top
        while(a.next!=None):
            a.print()
            a = a.next
        
        a.print()  
        
    def add_at(self, data, pos):
        node = Node(data) #new node to be added
        a = self.top     #counter starting from top
        while(pos-1!=0):
            a = a.next
            pos -= 1
        
        node.next = a.next
        a.next = node
        
    def delete_at(self, pos):
        a = self.top
        while(pos-1!=0):
            a = a.next
            pos -= 1
            
        a.next = a.next.next
        
        
        
    def kth(self,k):
        a = self.top
        count = 1
        while(count!=k):
            a = a.next
            count+=1
            
        while(a!=None):
            a.print()
            a = a.next
        
        
            
    
    
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.top = None
        self.end = None
        
    def add(self,data):
        node = Node(data)
        
        if self.top == self.end == None:
            self.top = self.end = node
        else:
            node.prev = self.end
            self.end.next = node
            self.end = node
            
            
    def print(self):
        a = self.top
        while(a!=None):
            a.print()
            a = a.next
        
#         a.print()  
        
        
        
    def dups(self):
        arr = []
        a = self.top
        while(a!=None):
            if a.data in arr:
                a.prev.next=a.next
                a.next.prev=a.prev
                a = a.next
                
            else:
                arr.append(a.data)
                a=a.next
                
       
        
        self.print()
            
        
    def partition(self,x):
        before_list = DoublyLinkedList()
        after_list = DoublyLinkedList()
        a = self.top
        while(a!=None):
        
            new = a
            if (new.data<x):
                before_list.add(new.data)
                
            else:
                after_list.add(new.data)
                
            a = a.next 


        before_list.end.next = after_list.top
        after_list.top.prev = before_list.end
        before_list.end = after_list.end
        return before_list
        
        
        
            
            
                    
                    
                    
                
    
    
```

```python
def loop_detection(ll):
    slow = ll.top
    fast = ll.top
    while(fast!=None and fast.next!=None):
        slow = slow.next
        fast = fast.next.next
        if slow==fast:
            break
            
            
    if fast==None or fast.next==None:
        return None
    
    slow = ll.top
    while (slow!=fast):
        slow = slow.next
        fast = fast.next
        
        
    return fast.data
    
    
```

```python
def simple_merge(A, B):
    temp=[]
    a_b = A+B
    i = 0
    s = len(A)
    t = len(A)+len(B)-1
    j = s
    while(i<=s-1 and j<=t):
        if(a_b[i]<a_b[j]):
            temp.append(a_b[i])
            i+=1
            
        else:
            temp.append(a_b[j])
            j+=1
            
    if i>s-1:
        for w in range(j,t+1):
            temp.append(a_b[w])
            
            
    else:
        for w in range(i,s):
            temp.append(a_b[w])
            
    return temp
            
    
```

```python
class Vertex:
    def __init__(self,x):
        self.id = x
        self.connectedTo = {}
        
        
    def addNeighbor(self,nbr, weight=0):
        self.connectedTo[nbr] = weight
        
        
    def getConnections(self):
        return self.connectedTo.keys()
    
    
    def getId(self):
        return self.id
    
    
    def getWeight(self,nbr):
        return self.connectedTo[nbr]
    
    def __str__(self):
        return str(self.id) + ' connected to: ' + str([x.id for x in self.connectedTo])
        
```

```python
class Graph:
    
    def __init__(self):
        self.vertList = {}
        self.numVertices = 0
        
        
        
    def addVertex(self, key):
        self.numVertices+=1
        newVertex = Vertex(key)
        self.vertList[key] = newVertex
        return newVertex
    
    
    def getVertex(self,n):
        if n in self.vertList:
            return self.vertList[n]
        
        else:
            
            return None
        
        
    def addEdge(self, f, t,cost=0):
        if f not in self.vertList:
            nv = self.addVertex(f)
            
        if t not in self.vertList:
            nv = self.addVertex(t)
            
            
        self.vertList[f].addNeighbor(self.vertList[t], cost)
        
        
        
        
    def getVertices(self):
        return self.vertList.keys()
    
    def __iter__(self):
        return iter(self.vertList.values())
    
    def __contains__(self,n):
        return n in self.vertList
            
        
            
```

```python
def buildGraph(wordList):
    d = {}
    g = Graph()
    for word in wordList:
        print(word)
        for i in range(len(word)):
            bucket = word[:i] + '_' + word[i+1:]
            if bucket in d:
                d[bucket].append(word)
                
            else:
                d[bucket] = [word]
                
                
                
    for bucket in d.keys():
        for word1 in d[bucket]:
            for word2 in d[bucket]:
                if word1 != word2:
                    g.addEdge(word1, word2,0)
                    
                    
    return g
```

```python
def bfs(s):
    level = {s:0}
    parent = {s:None}
    i = 1
    frontier = [s]
    while frontier:
        nextt=[]
        for u in frontier:
            for v in u.getConnections():
                if v not in level:
                    level[v] = i
                    parent[v] = u
                    nextt.append(v)
                    
                    
                    
        frontier = nextt
        i+=1
    
    return (level,parent)
```

```python
class LLNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
        
        
    def print(self):
        print("{}".format(self.data))
        
        
        
class SinglyLinkedList:
    def __init__(self):
        self.top = None
        self.end = None
        
    def add(self,data):
        node = LLNode(data)
        
        if self.top == self.end == None:
            self.top = self.end = node
        else:
            self.end.next = node
            self.end = node
       
```

```python

def copyBst(root, arr):
    if root==None:
        return
    copyBst(root.left, arr)
    arr.append(root.val)
    copyBst(root.right, arr)
    
    
def checkBst(root):
    arr = []
    copyBst(root, arr)
    if sorted(arr)!=arr:
        return False
    return True
```

```python
def update_bit(num, i, val):
    mask = ~(1<<i)
    return (num&mask)|(val<<i)
```

```python
def countEval(inp, ans):
    n = len(inp)
    if n==0:
        return 0
    if n==1:
        return bool(inp)
    count = 0
    for i in range(1,n,2):
        left = inp[:i]
        right = inp[i+1:]
        leftT = countEval(left, True)
        rightT = countEval(right, True)
        leftF = countEval(left, False)
        rightF = countEval(right, False)
        total = (leftT + leftF) * (rightT + rightF)
        total_true = 0
        if inp[i]=="^":
            print("here")
            total_true = leftT * rightF + leftF * rightT
        elif inp[i]=="&":
            total_true = leftT * rightT
            
        elif inp[i] == '|':
            total_true = leftT * rightT + leftF * rightT + leftT * rightF
        print(total_true, total)
        if ans:
            count+=total_true
        else:
            count+=(total - total_true)
            
    return count
        
```

```python
class Solution:
    # @param A : string
    # @return an integer
    def solve(self, A):
        dp = [[0 for i in range(len(A) + 1)] for j in range(len(A) + 1)]
        
        text1 = " " + A
        text2 = " " + A[::-1]
        
        for i in range(1, len(text1)):
            for j in range(1, len(text2)):
                if text1[i] == text2[j]:
                    dp[i][j] = 1 + dp[i-1][j-1]
                else:
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
            
        return dp[-1][-1]
```

```python
sol = Solution()
sol.solve("aniket")
```

```python
NO_OF_CHARS = 256
  
def badCharHeuristic(string, size): 
    ''' 
    The preprocessing function for 
    Boyer Moore's bad character heuristic 
    '''
  
    # Initialize all occurrence as -1 
    badChar = [-1]*NO_OF_CHARS 
  
    # Fill the actual value of last occurrence 
    for i in range(size): 
        badChar[ord(string[i])] = i; 
  
    # retun initialized list 
    return badChar 
  
def search(txt, pat): 
    ''' 
    A pattern searching function that uses Bad Character 
    Heuristic of Boyer Moore Algorithm 
    '''
    m = len(pat) 
    n = len(txt) 
  
    # create the bad character list by calling  
    # the preprocessing function badCharHeuristic() 
    # for given pattern 
    badChar = badCharHeuristic(pat, m)  
  
    # s is shift of the pattern with respect to text 
    s = 0
    while(s <= n-m): 
        j = m-1
  
        # Keep reducing index j of pattern while  
        # characters of pattern and text are matching 
        # at this shift s 
        while j>=0 and pat[j] == txt[s+j]: 
            j -= 1
  
        # If the pattern is present at current shift,  
        # then index j will become -1 after the above loop 
        if j<0: 
            return True
  
            '''     
                Shift the pattern so that the next character in text 
                      aligns with the last occurrence of it in pattern. 
                The condition s+m < n is necessary for the case when 
                   pattern occurs at the end of text 
               '''
            s += (m-badChar[ord(txt[s+m])] if s+m<n else 1) 
        else: 
            ''' 
               Shift the pattern so that the bad character in text 
               aligns with the last occurrence of it in pattern. The 
               max function is used to make sure that we get a positive 
               shift. We may get a negative shift if the last occurrence 
               of bad character in pattern is on the right side of the 
               current character. 
            '''
            s += max(1, j-badChar[ord(txt[s+j])]) 
inp1 = input()
inp2 = int(input())
arr = []
for i in range(inp2):
    arr.append([''.join(p) for p in list(permutations(input()))])
count = 0
for i in arr:
    for j in i:
        if j in inp1:
            count+=1
            break
            
if count==inp2:
    print('YES')
else:
    print('NO')
```
